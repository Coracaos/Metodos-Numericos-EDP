**2\. solución **

Reemplazamos en la solución aproximada para la ecuación Hiperbolica: 

$$u(h,2k) \cong \lambda u(0,k) + (2 - 2\lambda) u(h,k) + \lambda u(2h,k) - u(h,0)$$

usamos $u(h,k) = f(h) + kg(h)$ , $u(0,k) = 0$ y $u=f(h)$ por formulas dadas.

$$u(h,2k) \cong \lambda u(0,k) + (2 - 2\lambda) u(h,k) + \lambda u(2h,k) - u(h,0)$$

$$u(h,2k) \cong (2-2\lambda)(f(h) + kg(h)) + \lambda (f(2h) + kg(2h)) - f(h) $$

$$u(h,2k) \cong f(h) - 2\lambda f(h) + (2-2\lambda)kg(h) + \lambda f(2h) + \lambda kg(2h) $$

podemos notar que los valores de aproximación son diferentes por lo tanto la afirmación es falso.

**3\. solucion** 

Para este problema $\lambda = 0.45 \leq 0.5$, entonces debe ser una buena aproximación.

```python
import metodos as mt
import numpy as np
import pandas as pd

sol = mt.metodoExplicito(0.9, 1, 0.3, 0.1, 0.005, lambda x: np.sin(np.pi*x), lambda x: 0, lambda x: 0)
```

```python
def comparar(x, sol_aprox, sol_exac):
    
    error = abs(sol_exac - sol_aprox)

    error_por = abs(np.divide(error, sol_exac, 
                          out = np.zeros_like(error), 
                          where = np.round(sol_exac,6) != 0.0)*100)

    data = {"posición":x,
            "sol. exacta": np.round(sol_exac, 6),
            "sol. aprox.": np.round(sol_aprox, 6),
            "error": np.round(error, 6),
            "error %": np.round(error_por,2)}

    tabla = pd.DataFrame(data)
    tabla = tabla[["posición", "sol. exacta", "sol. aprox.", "error", "error %"] ]
    return tabla
```

```python
x = np.linspace(0, 1, 11)

sol_exac = np.exp(-0.9*(np.pi**2)*0.3) * np.sin(np.pi*x)

comparar(x, sol[-1], sol_exac)
```

Concluimos que es una buena aproximación ya que el error porcentual es de $3.74\%$ .

**8\. solución**

```python
def f(x):
    sol = np.zeros(len(x))
    
    lim = int((len(x) - 1)/2)
    
    sol[0:lim] = 2*x[0:lim]
    sol[lim:] = 2-2*x[lim:]
    return sol
```

```python
solutions = mt.metodoCrankNicolson(1, 1, 0.3, 0.05, 0.1, f, lambda x: 0, lambda x:0)
```

```python
t = np.linspace(0, 0.3, 4)
xs = np.linspace(0, 1, 21)
data = {"$t$" : t}

for x,sol in zip(xs, solutions.transpose()):
    data["x={0:.2f}".format(x)] = sol
    
tabla = pd.DataFrame(data)
tabla
```
