<div style="text-align:justify">
Las ecuaciones diferenciales parciales son necesarias para modelar ciertas cantidades que varian continuamente en más de un espacio, por ejemplo la temperatura varia continuamente tanto en el espacio como en el tiempo, entonces la temperatura puede ser modelado mediante una E.D.P, una de las ecuaciones que permite calcular la temperatura en un solido es la ecuación del calor. A continuación mediante un método numérico hallaremos las diferentes temperaturas con respecto al tiempo y el espacio para una dimension a partir de la ecuación del calor.
</div>

## E.D.P Parabolica - Ecuación del Calor para una Dimensión

<p>
<span id="eq01" style="float:right;">(1)</span>
$$
\begin{align*}    
\begin{cases}
 \dfrac{\partial u}{\partial t} = c^2 \dfrac{\partial^2 u}{\partial x^2} & \text{para } (x,t) \in (a,b)\times(0,\infty)\\ 
 {\scriptsize\mbox{ tal que}} & \\ 
 u(x,0) = f(x) & \text{para } x\in(a,b)    \\
 u(a,t) = 0 = u(b,t)  & \text{para } t\in(0,\infty)
\end{cases} 
\end{align*}
$$
</p>

<div style="text-align:justify;">
Donde $u(x,t)$ representa el valor de la temperatura en una barra para una posición $x$ y un tiempo $t$, ademas $a$ y $b$ son los extremos de la barra.
</div>

## Método de Diferencias Finitas

<div style="text-align:justify">
El Método de Diferencias Finitas nos permite aproximar valores de derivadas parciales de una forma sencilla a partir de unas formulas que pueden ser deducidas por el polinomio de taylor. Las formulas para aproximar las derivadas parciales de la ecuación (<a style="text-decoration:none;" href="#eq01">1</a>) son las siguientes:
</div>

$u = u(x,y)$

<p>
<span id="eq02" style="float:right;">(2)</span>
$$
\frac{\partial u(x_i,t_j)}{\partial t} \simeq \frac{u(x_i, t_j + h) - u(x_i, t_j)}{h}
$$
</p>

<p>
<span id="eq03" style="float:right;">(3)</span>
$$
\frac{\partial u(x_i,t_j)}{\partial t} \simeq \frac{u(x_i, t_j) - u(x_i, t_j - h)}{h}
$$
</p>

<p>
<span id="eq04" style="float:right;">(4)</span>
$$
\frac{\partial^2 u(x_i,t_j)}{\partial t^2} \simeq \frac{u(x_i,t_{j+1}) - 2u(x_i,t_j) + u(x_i,t_{j-1})}{h^2}
$$
</p>

<p>
<span id="eq05" style="float:right;">(5)</span>
$$
\frac{\partial^2 u(x_i,t_j)}{\partial x^2} \simeq \frac{u(x_{i+1},t_j) - 2u(x_i,t_j) + u(x_{i-1},t_j)}{h^2}
$$
</p>

<div style="text-align:justify">
ahora vamos realizar una partición regular : $ h = \dfrac{b - a}{m}$ tamaño de paso 
</div>

$$ u(x_i,0) = f(x_i) $$

<p>
<span id="eq06" style="float:right;">(6)</span>
$$
\frac{\partial u(x_i,t_j)}{\partial t} \simeq \frac{ u(x_i,t_{j+1}) - u(x_i,t_j) }{ k } 
$$

<p>
<span id="eq07" style="float:right;">(7)</span>
$$
\frac{\partial^2 u(x_i, t_j)}{\partial x^2} \simeq \frac{ u(x_{i+1},t_j) - 2u(x_i,t_j) + u(x_{i-1},t_j) }{ h^2 }
$$
</p>

$u(x_i,t_j) \simeq w_{i,j}$

$w_{i,j}$ denota la aproximación a $u(x_i,t_j)$ en el instante $t_j$ y en el punto $x_i$.

utilizamos $w_{i,j}$ como valor aproximado de $u(x_i,t_j)$ en la ecuación del calor y reemplazamos los valores de las derivadas parciales por sus aproximaciones dadas en las ecuaciones (<a style="text-decoration:none;" href="#eq06">6</a>) y (<a style="text-decoration:none;" href="#eq07">7</a>)

<p>
<span id="eq08" style="float:right;">(8)</span>
$$
\frac{w_{i,j+1} - w_{i,j}}{k} = c^2 \frac{w_{i+1,j} - 2w_{i,j} + w_{i-1,j}}{h^2}
$$
</p>

<div style="text-align:justify">   
denotamos a $\lambda$ como :
</div>

<p>
<span id="eq09" style="float:right;">(9)</span>
$$
\lambda = \dfrac{c^2 k}{h^2} > 0
$$
</p>


entonces en la ecuació <a style="text-decoration:none;" href="#eq08">(8)</a> obtenemos : 

<p>
<span id="eq10" style="float:right;">(10)</span>
$$
w_{i,j+1} = \lambda w_{i-1, j} + (1 - 2 \lambda) w_{i, j} + \lambda w_{i+1, j}
$$
</p>


<div style="text-align:justify">
utilizando esta formula para $i=1,2,3,..,n-1$ obtenemos
</div>
<p>
$$w_{1,j+1} = \lambda w_{0,j} + (1-2\lambda)w_{1,j} + \lambda w_{2,j}$$
$$w_{2,j+1} = \lambda w_{1,j} + (1-2\lambda)w_{2,j} + \lambda w_{3,j}$$
$$w_{3,j+1} = \lambda w_{2,j} + (1-2\lambda)w_{3,j} + \lambda w_{4,j}$$
$$\vdots $$
</p>




expresando esto matricialmente obtenemos

$$ 
\left( \begin{array}{c}
        w_{1,j+1}  \\
        w_{2,j+1}  \\
        w_{3,j+1}  \\
        \vdots   \\
        w_{n-1,j+1}\\
       \end{array}
\right)  = \left( \begin{array}{ccccc}
        1-2\lambda & \lambda    & 0          & \cdots  & 0          \\
        \lambda    & 1-2\lambda & \lambda    & \cdots  & 0          \\
        0          & \lambda    & 1-2\lambda & \cdots  & 0          \\
        \vdots     & \vdots     & \vdots     & \ddots  & \lambda   \\
        0          & 0          & 0          & \lambda & 1-2\lambda \\        
        \end{array}
\right) \left( \begin{array}{c}
       w_{1,j}  \\
       w_{2,j}  \\
       w_{3,j}  \\
       \vdots     \\
       w_{n-1,j}\\
       \end{array}
\right)       
$$

<p>
<span id="eq11" style="float:right;">(11)</span>
$$

w^{(j)} = A w^{(j-1)}

$$
</p>

### Ejercicio 01

Hallar la temperatura para $t = 0.3$ s. de una barra de un metro cuyos extremos se mantienen a $20°$C y $40°$C la temperatura inicial de la barra es de $100°$C y el coeficiente $a = 0.1$. Tomar $\Delta x = 0.2$ m. y $\Delta t = 0.1$ s. justificar la aplicabilidad del método explicito.

#### solución

Para justificar la aplicabilidad de este método debemos comprobar que el parametro de la ecuación <a style="text-decoration:none;" href="#eq08">(8)</a> sea:

<p>
$$

\lambda = \dfrac{c^2 k}{h^2} \leq 0.5

$$
</p>

comprobemos esto con los datos del problema:

$$
\lambda = \dfrac{(0.1) (0.1)}{0.04} = 0.25 \leq 0.5 
$$

por lo tanto podemos aplicar este metodo. A continuación se muestra el codigo :

Solución mediante el uso de la ecuación <a style="text-decoration:none;" href="#eq08">(10)</a>

```python
import numpy as np

def metodoExplicito(c, L, T, h, k, f, a, b):
    
    """
    método numérico para resolver la ecuación del calor
    
    parametros
    ----------
    c: parametro de la ecuacion del calor
    L: longitud de la barra
    T: tiempo final
    h: paso para la partición de la longitud 
    k: paso para la partición del tiempo  
    f: funcion para los valores iniciales de temperatura
    a: funcion para la temperatura en el extremo izquiero
    b: funcion para la temperatura en el extremo derecho
    
    retorna una matriz con las soluciones de la temperatura para cada valor del tiempo
    """
    
        
    r = k*c/h**2 # calculamos el valor de r (parametro de la disusion termica) 
    
    m = round(L/h) + 1 #m puntos a lo largo de la barra
    n = round(T/k) + 1 #n puntos en el tiempo
    
    x = np.linspace(0, L, m) #particion en la longitud de la barra
    
    sol = np.zeros((n,m)) # inicializamos la matriz de solucion
        
    sol[0] = f(x) # valor inicial de temperatura
    
    sol[:,0] = a(0) #valor de frontera extremo derecho
    sol[:,-1] = b(L) #valor de frontera extremo izquierdo
    
    #aplicamos la formula 
    for j in range(n-1):
        for i in range(1,m-1):
            sol[j+1,i] = r*sol[j,i+1] + (1-2*r)*sol[j,i] + r*sol[j,i-1]
        
    return sol
```

```python
solutions = metodoExplicito(0.1, 1, 0.3, 0.2, 0.1, lambda x: 100, lambda x: 20, lambda x: 40)
solutions
```

Ahora graficamos estas soluciones con el siguiente codigo :

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(7.0,4.0), dpi=100)
ax = fig.add_subplot(1,1,1)

ts = np.linspace(0,0.3,4)

for sol,t in zip(solutions,ts):
    ax.plot(np.linspace(0,1,6),sol, label="$t = {0:.1f}$".format(t))
    
ax.grid(linestyle=':')
ax.set_title("Temperatura en la Barra")
ax.set_xlabel("Ubicación en la barra")
ax.set_ylabel("Temperatura")
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
    
```

Solución mediante el uso de la ecuación  <a style="text-decoration:none;" href="#eq11">(11)</a>.

```python
def tridiag(a,b,c,N):
    A = np.zeros((N,N))
    
    np.fill_diagonal(A[:-1,1:],a)
    np.fill_diagonal(A,b)
    np.fill_diagonal(A[1:,:-1],c)
    
    return A


def metodoExplicitoMatriz(c, L, T, h, k, f, a, b):
    
    r = c*k/h**2
    m = round(L/h) + 1
    n = round(T/k) + 1
    
    x = np.linspace(0, L, m)
    
    sol = np.zeros((n,m))
    sol[0] = f(x)
    sol[:,0] = a(0) 
    sol[:,-1] = b(L)
    
    A = tridiag(r, 1-2*r, r, m-2) 
    
    for i in range(n-1):
        sol[i+1,1:-1] = np.dot(A, sol[i,1:-1])
        sol[i+1,1] += r*sol[i,0]
        sol[i+1,-2] += r*sol[i,-1]
        
    return sol
     
```

```python
solutions = metodoExplicitoMatriz(0.1, 1, 0.3, 0.2, 0.1, lambda x: 100, lambda x: 20, lambda x: 40)
```

### Ejercicio 02

Use los siguientes tamaños de paso $\Delta x = 0.1$ y $\Delta t = 0.0005$ para aproximar la solución de la ecuación :

$$\dfrac{\partial u}{\partial t}(x,t) - \dfrac{\partial^2 u}{\partial x^2}(x,t) = 0, \quad 0<x<1, \quad 0\leq t$$

con las condiciones de contorno :

$$ u(0,t) = u(1,t) = 0$$

y las condiciones iniciales :

$$u(x,0) = \sin(\pi x)$$

#### solución

Usamos la función creada en el ejercicio anterior

```python
solutions = metodoExplicito(1, 1, 0.5, 0.1, 0.0005, lambda x: np.sin(np.pi*x), lambda x: 0, lambda x: 0)
```

Ahora graficamos la solución

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

#preparamos los datos
x = np.linspace(0, 1, round(1/0.1) + 1)
t = np.linspace(0, 0.5, round(0.5/0.0005) + 1)
x,t = np.meshgrid(x,t)

#iniciamos nuestra figura
fig = plt.figure(figsize=(8.0,5.0), dpi=120)
ax = fig.add_subplot(111, projection="3d")

#creamos el trazado de la superficie
surf = ax.plot_surface(x,t,solutions, cmap=cm.coolwarm)

#propiedades del grafico
ax.set_title("Variación de la temperatura")
ax.set_xlabel("ubición en la barra", size = 8)
ax.set_ylabel("tiempo", size = 8)

#inicalizamos los angulos para visualizar la imagen
ax.view_init(30, 110)

#agregamos la barra de color
fig.colorbar(surf, shrink=0.5, aspect=10, ticks=np.linspace(0,1,6), label="temperatura")

plt.show()
```
